```python
np.random.seed(42)
x = np.random.randn(100)

# Рассчитываем гистограмму в ручную
bins = np.linspace(-5, 5, 20)
counts = np.zeros_like(bins)

# Ищем подходящий интервал для каждого x
i = np.searchsorted(bins, x)

# Добавляем 1 к каждому из интервалов
np.add.at(counts, i, 1)

# Визуализируем результаты
plt.plot(bins, counts, drawstyle='steps');
```

```python
plt.hist(x, bins, histtype='step');
```

```python
print("NumPy routine:")
%timeit counts, edges = np.histogram(x, bins)  # histogram()
print("Custom routine:")
%timeit np.add.at(counts, np.searchsorted(bins, x), 1)
```

```python
x = np.random.randn(1000000)
print("NumPy routine:")
%timeit counts, edges = np.histogram(x, bins)
print("Custom routine:")
%timeit np.add.at(counts, np.searchsorted(bins, x), 1)
```

```python
X = rand.rand(10, 2)

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # Plot styling
plt.scatter(X[:, 0], X[:, 1], s=100);
```

```python
dist_sq = np.sum((X[:,np.newaxis,:] - X[np.newaxis,:,:]) ** 2, axis=-1)
```

```python
# Для каждой пары точек вычисляем разности их координат
differences = X[:,np.newaxis,:] - X[np.newaxis,:,:]
differences.shape
```

```python
# Возводим разности координат в квадрат
sq_differences = differences ** 2
sq_differences.shape
```

```python
# Суммируем квадраты разностей координат 
# для получения квадрата растояния
dist_sq = sq_differences.sum(-1)
dist_sq.shape
```

```python
# проверка
dist_sq.diagonal()
```

```python
nearest = np.argsort(dist_sq, axis=1)
print(nearest)
```

```python
K = 2
nearest_partition = np.argpartition(dist_sq, K + 1, axis=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=100)

# Рисуем линии из каждой точки к ее двум ближайшим соседям
K = 2

for i in range(X.shape[0]):
    for j in nearest_partition[i, :K+1]:
        # Чертим линию от X[i] до X[j]
        # Используем магическую функцию zip:
        plt.plot(*zip(X[j], X[i]), color='black')
```

```python
import numpy as np

name = ['Alice', 'Bob', 'Cathy', 'Doug']
age = ['25', '45', '37', '19']
weight = [55.0, 85.5, 68.0, 61.5]
```
