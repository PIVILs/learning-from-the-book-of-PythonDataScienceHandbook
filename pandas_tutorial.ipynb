```python
def make_df(cols, ind):
    """ Быстро создаем объект DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
           for c in cols}
    return pd.DataFrame(data, ind)

# Экземпляр DataFrame
make_df('ABC', range(3))
```

Увеличение производительности библиотеки Pandas: eval() and query()

Основания для использования функций query() and eval(): составные выражения

```python
rng = np.random.RandomState(42)
x = rng.rand(int(1e6)) #1_000_000
y = rng.rand(int(1e6))
%timeit x + y
```

```python
%timeit np.fromiter((xi + yi for xi, yi in zip(x, y)), dtype=x.dtype, count=len(x))
```

```python
mask = (x > 0.5) & (y < 0.5)
```

Эквивалентно:
tmp1 = (x > 0.5)
tmp2 = (y > 0.5)
mask = tmp1 & tmp2

Для каждого промежуточного шага явным образом выделяется оперативная память

Библиотека Numexpr позволяет вычислить подобные составные выражения поэлементно, не требуя выделения памяти под промежуточные массивы целиком.
Принимает на входе строку

Numexpr вычисляет выражение, не используя полноразмерных врЕменных массивов, а потому оказывается намного более эффективной, чем NumPy

```python
import numexpr
mask_numexpr = numexpr.evaluate('(x > 0.5) & (y < 0.5)')
np.allclose(mask, mask_numexpr)
```

Use func pandas.eval()

Функция eval() применяет строковые выражения для эффективных вычислительных операций с объектами DataFrame

```python
nrows, ncols = 100000, 100
rng = np.random.RandomState(42)
df1, df2, df3, df4 = (pd.DataFrame(rng.rand(nrows, ncols)) for i in range(4))
```

```python
%timeit df1 + df2 + df3 + df4
```

```python
%timeit pd.eval('df1 + df2 + df3 + df4') #использует работает на 50% быстрее и много меньше памяти
```

```python
np.allclose(df1 + df2 + df3 + df4, eval('df1 + df2 + df3 + df4'))
```

Поддерживаемые фунцией pd.eval() операции

```python
df1, df2, df3, df4, df5 = (pd.DataFrame(rng.randint(0, 1000, (100, 3))) for i in range(5))
```

Арифметические функции

```python
result1 = -df1 * df2 / (df3 + df4) - df5
result2 = pd.eval('-df1 * df2 / (df3 + df4) - df5')
np.allclose(result1, result2)
```

Операторы сравнения

```python
result1 = (df1 < df2) & (df2 <= df3) & (df3 != df4)
result2 = pd.eval('df1 < df2 <= df3 != df4')
np.allclose(result1, result2)
```

Побитовые операторы

```python
result1 = (df1 < 0.5) & (df2 < 0.5) | (df3 < df4)
result2 = pd.eval('(df1 < 0.5) & (df2 < 0.5) | (df3 < df4)')
np.allclose(result1, result2)
```

```python
result3 = pd.eval('(df1 < 0.5) and (df2 < 0.5) or (df3 < df4)')
np.allclose(result1, result3)
```

доступ к атрутам obj.attr 
доступ к индексам obj[index]

Локальные переменные в методе DataFrame.eval()

```python
column_mean = df.mean(1)
result1 = df['A'] + column_mean
result2 = df.eval('A + @column_mean')
#Символ @ отмечает имя переменной, а не имя столбца. 
#Символ @ только для DataFrame.eval()
np.allclose(result1, result2)
```

#### Метод DataFrame.query() 

```python
result1 = df[(df.A < 0.5) & (df.B < 0.5)]
result2 = pd.eval('df[(df.A < 0.5) & (df.B < 0.5)]')
np.allclose(result1, result2)
```

```python
result2 = df.query('A < 0.5 and B < 0.5')
np.allclose(result1, result2)
```

```python
Cmean = df['C'].mean()
result1 = df[(df.A < Cmean) & (df.B < Cmean)]
result2 = df.query('A < @Cmean and B < @Cmean')
np.allclose(result1, result2)
```
